# Import pandas, matplotlib, and data file

# dropping columns from our data set

```python
dfcrime = crime.drop(labels = ['modus_operandi', 'victim_sex', 'victim_descent', 
                              'weapon_code', 'weapon_description', 'crime_code_2', 'crime_code_3', 'crime_code_4'], axis = 1)
```

```python
dfcrime1 = dfcrime
```

```python
dfcrime1
```

# converting object data type to data data type

```python
dfcrime1.dtypes
```

```python
dfcrime1['date_reported'] = pd.to_datetime(dfcrime1['date_reported'], format = '%Y-%m-%d')
```

```python
dfcrime1['date_occurred'] = pd.to_datetime(dfcrime1['date_occurred'], format= '%Y-%m-%d %H:%M:%S')
```

```python
dfcrime1.dtypes
```

```python
dfcrime1
```

```python
dfcrime1.describe()
```

# describing only specific columns with include

```python
dfcrime1.describe(include = ['float', 'int'])
```

# counts of values in specific columns

```python
dfcrime1['victim_age'].value_counts().reset_index()
```

# cleaning data: filtering out unrealistic values

```python
dfcrime_age = dfcrime1[dfcrime1['victim_age'] > 0]
```

```python
dfcrime_age['victim_age'].value_counts()
```

```python
dfcrime_age
```

```python
dfcrime1
```

# inserting columns to data set. adding columns for year, month, day

```python
dfcrime1['year_occurred'] = dfcrime1['date_occurred'].dt.year
dfcrime1['month_occurred'] = dfcrime1['date_occurred'].dt.month
dfcrime1['day_occurred'] = dfcrime1['date_occurred'].dt.day
```

```python
dfcrime1
```

# quick visualization

```python
dfcrime1['month_occurred'].value_counts().reset_index().sort_values(by = 'month_occurred').plot(x = 'month_occurred', y = 'count')
```

```python
dfcrime1['day_occurred'].value_counts().reset_index().sort_values(by = "day_occurred").plot(x = 'day_occurred', y ='count')
```

```python
dfcrime1['year_occurred'].value_counts().reset_index().sort_values(by = "year_occurred").plot(x = 'year_occurred', y ='count', kind = 'bar')
```

```python
plt.hist(dfcrime_age['victim_age'], edgecolor = 'black', bins = 20)
```

```python
dfcrime2 = dfcrime1[~((dfcrime1['month_occurred'] == 12) & (dfcrime1['year_occurred'] == 2023))]
```

```python
dfcrime2
```

```python
dfcrime2.groupby(['year_occurred', 'month_occurred']).size()
```

```python
dfcrime2.groupby(['year_occurred', 'month_occurred']).size()
```

```python
dfcrime
```

```python
dfcrime2['date_occurred_notime'] = dfcrime2['date_occurred'].dt.date
```

```python
dfcrime2
```

```python
dfcrime2.groupby(['date_occurred_notime', 'crime_description']).size()
```

```python
crime_per_day = dfcrime2.groupby(['date_occurred_notime', 'crime_description']).size()
```

```python
crime_per_day.groupby('crime_description').mean().sort_values(ascending = False).head(20).reset_index(name = 'average_daily_crime')
```

```python
crime_per_day1 = crime_per_day.groupby('crime_description').mean().sort_values(ascending = False).head(20).reset_index(name = 'average_daily_crime')
```

```python
plt.barh(crime_per_day1['crime_description'], crime_per_day1['average_daily_crime'])
```
